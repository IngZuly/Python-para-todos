```python
# --- Sistema de Alerta de Parámetros ---

# 1. Solicitar al usuario el nombre del área/parámetro
area_parametro = input("Ingrese el nombre del área o parámetro (ej. Temperatura del reactor): ")

# 2. Solicitar al usuario el valor crítico actual
try:
    valor_critico = float(input(f"Ingrese el valor actual de '{area_parametro}': "))
except ValueError:
    print("Entrada inválida. Por favor, ingrese un número para el valor crítico.")
    exit()

# 3. Solicitar al usuario el límite
try:
    limite = float(input(f"Ingrese el valor límite para '{area_parametro}': "))
except ValueError:
    print("Entrada inválida. Por favor, ingrese un número para el límite.")
    exit()

# 4. Evaluar la condición y mostrar la alerta o el estado estable
print("\n--- Resultado ---")
if valor_critico > limite:
    print(f"¡ALERTA! El valor de '{area_parametro}' ({valor_critico}) supera el límite ({limite}).")
else:
    print(f"ESTABLE. El valor de '{area_parametro}' ({valor_critico}) está dentro del límite ({limite}).")
```
